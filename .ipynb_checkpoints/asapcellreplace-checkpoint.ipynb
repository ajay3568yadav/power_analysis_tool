```python
import linecache as lc
import re

def extract_key(string_line):
    li = string_line.split()
    if len(li) >= 3:
        li[1] = li[1].replace("(", "")
        li[1] = li[1].replace(")", "")
    return li[1]

def extract(string_line):
    li = re.findall(r'\b(?<!when\s)\w+\b', string_line)
    if "when" in li:
        li.remove("when")
    return li

pattern = "cell ("
cell_map_AO = {}
cell_map_INVBUF = {}
cell_map_OA = {}
cell_map_SIM = {}
cell_map_SEQ = {}

def cell_mapping(filename, mapname, portlinenumber):
    line_num_x = 0
    try:
        with open(filename, "r") as f:
            line = f.readline()
            line_num_x += 1
            while line:
                if line.find(pattern) >= 0:
                    mapname[extract_key(line.strip())] = extract(lc.getline(filename, line_num_x + portlinenumber).strip())
                line = f.readline()
                line_num_x += 1
    except FileNotFoundError:
        print("File not found.")
    except IOError:
        print("Error reading the file.")
    print(mapname)
```

```python
print("#########cell map for AO###########")
cell_mapping("asap7sc7p5t_AO_LVT_TT_nldm_211120.lib",cell_map_AO,12)
print("#########cell map for OA###########")
cell_mapping("asap7sc7p5t_OA_LVT_TT_nldm_211120 (1).lib",cell_map_OA,12)
print("#########cell map for SIMPLE###########")
cell_mapping("asap7sc7p5t_SIMPLE_LVT_TT_nldm_211120.lib",cell_map_SIM,14)
print("#########cell map for SEQ###########")
cell_mapping("asap7sc7p5t_SEQ_LVT_TT_nldm_220123.lib",cell_map_SEQ,12)
print("#########cell map for INVBUF###########")
cell_mapping("asap7sc7p5t_INVBUF_SLVT_TT_nldm_220122.lib",cell_map_INVBUF,14)

keylistAO=list(cell_map_AO.keys())
keylistOA=list(cell_map_OA.keys())
keylistSIM=list(cell_map_SIM.keys())
keylistSEQ=list(cell_map_SEQ.keys())
keylistINVBUF=list(cell_map_INVBUF.keys())

finalkeylist=keylistAO+keylistOA+keylistSIM+keylistSEQ+keylistINVBUF
#print(finalkeylist)
```

```python
def get_characters_until_x(string):
    i = string.find("x")
    return string[:i]
```

```python
dictlist = [cell_map_AO, cell_map_OA, cell_map_SIM, cell_map_SEQ, cell_map_INVBUF]
def getAbvlist(abvlist:list,dictlist:list):
    for maps in dictlist:
        keylist=list(maps.keys())
        for key in keylist:
            abv=get_characters_until_x(key)
            if abv in abvlist:
                continue
            abvlist.append(abv)
```

```python
abvlist=[]
getAbvlist(abvlist,dictlist)
print(abvlist)
```

```python
def replace_with_asap(line:str,map_dictionary:dict):
    dummyline=line.strip()
    if dummyline[:4]=="GEN_":
        dummylinelist=dummyline.split()
        if dummylinelist[0] in list(map_dictionary.keys()):
            asapcell=map_dictionary[dummylinelist[0]]
            line=line.replace(dummylinelist[0],asapcell)
    return line
```

```python
with open("ExampleRocketSystem_GEN.txt","w") as writer:
    with open("ExampleRocketSystem_GEN.gv","r") as reader:
        lines=reader.readline()
        while lines:
            writer.writelines(replace_with_asap(lines,map_dictionary))
            #lines.replace(".zn(",",.Yn(")
            #lines.replace(".z(",",.Y(")
            #lines.replace(".zn(",",.Yn(")
            #lines.replace(".zn(",",.Yn(")
            #lines.replace(".zn(",",.Yn(")
            #lines.replace(".zn(",",.Yn(")
            lines=reader.readline()
        writer.close()
```

```python
class OutputPair:
    def __init__(self,cellname,output):
        self.cellname=cellname
        self.output=output
def extract_strings(filename):
    extracted_strings = []
    outputlist=[]
    key=""
    line = " "
    with open(filename, 'r') as file:
        while line:
            if line.find("ASAP7_75t")>=0:
                key=line.strip().split()[0]
                print(key)
            zindex=line.find('.z(')
            znindex=line.find('.zn(')
            qindex=line.find('.q(')
            dindex= line.find('.dout(')
            
            start_index=max(zindex,znindex,qindex,dindex)
            end_index=line.find(') );')
            if start_index != -1 and end_index == -1:
                start_index = 0
                while end_index == -1:
                    line=file.readline()
                    end_index = line.find(') );')
                start_index = 0
            if start_index != -1 and end_index != -1:
                extracted_string = line[start_index + 2:end_index].strip()
                extracted_string=extracted_string.replace("n(","")
                extracted_string=extracted_string.replace("(","")
                extracted_strings.append(extracted_string)
                outputlist.append(OutputPair(key,extracted_string))
            line=file.readline()
    
    return [extracted_strings,outputlist]

filename = 'ExampleRocketSystem_GEN.gv' 

outputlist_tuple = extract_strings(filename)
print(outputlist_tuple[0])
```
