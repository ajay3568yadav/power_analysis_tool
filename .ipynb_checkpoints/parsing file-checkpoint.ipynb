```python
time=""
def process_line(f, objectdict, time):
    line = f.readline().rstrip("\n")
    while line != "":
        if line[0] == "#":
            time = line[1:]
        elif line[0] == "b" or line[0] == "B":
            l1 = line.split()
            objectdict[l1[-1]].toggle[time] = l1[0]
        else:
            objectdict[line[1:]].toggle[time] = line[0]
        line = f.readline().rstrip("\n")

process_line(f, objectdict, time)
```
